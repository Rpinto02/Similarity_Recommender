```python
import pandas as pd
import matplotlib.pylab as plt 
import numpy as np
import seaborn as sns
import plotly.graph_objs as go
from plotly.offline import iplot
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from scipy import stats
import umap as umap
import gower


def load_data(filepath):
    '''takes in the file path where the data is store and returns a pandas dataframe.
    filepath must be entered as a string
    '''
    
    df = pd.read_csv(filepath)
    return df;

filepath = '../data/market_ETL.csv'
market_ID = load_data(filepath)

market = market_ID.drop(columns=['id'])
market.head()
```

```python
market.select_dtypes(exclude='int64')
```

```python
var_cont = market['idade_empresa_anos'].values.reshape(-1,1)
var_cont
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
var_cont_standarized = sc.fit_transform(var_cont)


var_cont_standarized
```

```python
standarized = market
standarized['idade_empresa_anos'] = var_cont_standarized
standarized
```

```python
dissimilarity_matrix = gower.gower_matrix(standarized)
dissimilarity_matrix
```

```python
dissimilarity_matrix.shape
```

```python
np.percentile(dissimilarity_matrix,25), dissimilarity_matrix.min(),dissimilarity_matrix.max()
```

```python
eps_dist = dissimilarity_matrix[0]
plt.hist(eps_dist,bins=30)
plt.ylabel('n');
plt.xlabel('Epsilon distance')
```

```python
d = dissimilarity_matrix[0]
plt.plot(d)
plt.ylabel("k-distances")
plt.grid(True)
plt.show()
```
