The purpose of this notebook is simply to generate a database sample so we can use in the streamlit app to showcase our engine.

```python
import pandas as pd
import numpy as np



def load_data(filepath):
    '''takes in the file path where the data is store and returns a pandas dataframe.
    filepath must be entered as a string
    '''
    
    df = pd.read_csv(filepath)
    return df;

filepath = '../app/data/market_ETL.csv'
market_ID = load_data(filepath)

portfolio1 = pd.read_csv('../app/data/portfolio1_ETL.csv')
portfolio2 = pd.read_csv('../app/data/portfolio2_ETL.csv')
portfolio3 = pd.read_csv('../app/data/portfolio3_ETL.csv')

#adding 10k entries in order not to slow the app when we open it.
market_sample = market_ID[0:10000].append(portfolio1).append(portfolio2).append(portfolio3)
```
